```python
sns.set(rc = {'figure.figsize':(15,6)})

sns.barplot(x="Category", y="Calories", data=mcdon_df)
plt.show()


sns.barplot(x="Category", y="Total Fat (% Daily Value)", data=mcdon_df)
plt.show()


sns.barplot(x="Category", y="Cholesterol (% Daily Value)", data=mcdon_df)
plt.show()


sns.barplot(x="Category", y="Sodium (% Daily Value)", data=mcdon_df)
plt.show()



sns.barplot(x="Category", y="Carbohydrates (% Daily Value)", data=mcdon_df)
plt.show()


sns.barplot(x="Category", y="Dietary Fiber (% Daily Value)", data=mcdon_df)
plt.show()


sns.barplot(x="Category", y="Sugars", data=mcdon_df)
plt.show()


sns.barplot(x="Category", y="Protein", data=mcdon_df)
plt.show()
```

```python
data_calory = mcdon_df.sort_values(by = ['Calories'], ascending=False)
sns.barplot(y="Item", x="Calories", data = data_calory.head(10))
```

```python
data_Protein = mcdon_df.sort_values(by = ['Protein'], ascending=False)
sns.barplot(y="Item", x="Protein", data = data_Protein.head(10))
```

```python
data_Fat = mcdon_df.sort_values(by = ['Total Fat (% Daily Value)'], ascending=False)
sns.barplot(y="Item", x="Total Fat (% Daily Value)", data = data_Fat.head(10))
```
